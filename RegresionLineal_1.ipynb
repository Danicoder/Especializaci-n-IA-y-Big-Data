#REGRESION LINEAL

```
#Predecimos el coste de inscidente de seguridad en base al número de equipos que ha sido afectado.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Incidentes
x = 2 * np.random.rand(100,1)
#coste de los incidentes
y = 4 + 3 * x + np.random.rand(100,1)
print("La longitud del conjunto de datos es:", len(x), "incidentes pasados \n")

#crea la figura y los ejes
fig, ax = plt.subplots()
# Dibujar puntos
ax.plot(x,y,"b.") # color azul (b) y a puntitos (.)
#Etiquetas
plt.xlabel("Equipos Afectados (u/1000)")
plt.ylabel("Coste de incidente ((u/10000)")
#Visualización
plt.show()

#Modificamos el conjunto de datos
Data={"n_equipos_afectados":x.flatten(),"coste":y.flatten()}
#Convertimos los datos a Dataframe
df=pd.DataFrame(Data)
#visualizamos los 10 primeros
df.head(10)

#Transformo los datos en valores reales (eje X)
df["n_equipos_afectados"] = df["n_equipos_afectados"]*1000
df["n_equipos_afectados"] = df["n_equipos_afectados"].astype("int")
df.head(10)

#Transformo los datos en valores reales (eje Y)
df["coste"] = df["coste"]*1000
df["coste"] = df["coste"].astype("int")
df.head(10)

#Representación gráfica con valores reales
#crea la figura y los ejes
fig, ax = plt.subplots()
# Dibujar puntos
ax.plot(x,y,"b.") # color azul (b) y a puntitos (.)
#Etiquetas
plt.xlabel("Equipos Afectados")
plt.ylabel("Coste de incidente")
#Visualización
plt.show()

#Construcción de MODELO (aprendizaje supervisado)
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df["n_equipos_afectados"].values.reshape(-1,1),df["coste"].values)

#Ver el valor aproximado de tetha 0 (0^0)
print(f"Valor de Tetha 0: {lin_reg.intercept_}")
#Ver el valor aproximado de tetha 1 (0^1)
print(f"Valor de Tetha 1: {lin_reg.coef_ }")
#valor mínimo y máximo
x_min_max = np.array([[df['n_equipos_afectados'].min()], [df['n_equipos_afectados'].max()]])
y_train_pred = lin_reg.predict(x_min_max)
print(f"Valor mínimo y máximo: {y_train_pred}")
```
