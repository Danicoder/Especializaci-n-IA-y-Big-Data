```
#Librerías pandas,numpy, seaborn y matploit
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Leer el archivo CSV
grafo = pd.read_csv('anscombe.csv')
#Resumen estadístico
grafo.describe()
```

```
#Máximo y el mínimo con el método agg
print(grafo.agg(['max','min']))
#Media y varianza de cada conjunto de datos con el método agg(agragate)
print(grafo.groupby(['x','y']).agg({'x': ['mean','var'],
                              'y': ['mean','var'],}))
#Histograma
plt.hist('x')
plt.hist('y')
plt.title('Histograma de medias y varianzas de x e y')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```
