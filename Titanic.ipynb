```
print("**********************EJERCICIO 1**********************")
import pandas as pd
##LECTURA DEL ARCHIVO
df = pd.read_csv('titanic_train.csv')
print(df)
print("**********************EJERCICIO 2**********************")
#Imprimir 20 primeras filas
print(df.head(20))
print("**********************EJERCICIO 3**********************")
#Imprimir las últimas 5 filas
print(df.tail(5))
print("**********************EJERCICIO 4**********************")
#Imprime la información del fichero para saber el tipo de dato
print(df.info())
print("**********************EJERCICIO 5**********************")
#Imprime el número de filas y columnas
print(f"número de filas y columnas: {df.shape}")
print("**********************EJERCICIO 6**********************")
#Imprime resumen estadístico
print(df.describe())
print("**********************EJERCICIO 7**********************")
titanic= pd.read_csv("titanic_train.csv")
print(titanic.describe())
#Imprimir datos con pasajeros > 30 años
pasajeros_mayor_30 = titanic[titanic['Age']>30].describe()
print(f" Pasajeros mayores de 30 años: {pasajeros_mayor_30}")
quince_primeros = pasajeros_mayor_30.head(15)
print(f" 15 primeros de la lista anterior: {quince_primeros}")
#Imprimir pasajeros con edad comprendida entre 45 y 55 años
pasajeros_entre_45_55= titanic[(titanic['Age']>=45) & (titanic['Age']<=55)].describe()
print(f" pasajeros con edad comprendida entre 45 y 55 años: {pasajeros_entre_45_55}")

#Para saber la cantidad de pasajeros que cumplen la condición
print("**********************EJERCICIO 8**********************")
cantidad = pasajeros_entre_45_55['PassengerId'].count()
print(f"Cantidad de pasajeros: {cantidad}")

#Pasajeros que viajan solos
print("**********************EJERCICIO 9**********************")
pasajeros_singles= titanic[titanic['SibSp']==0].shape[0]
print(f"Pasajeros que viajan solos: {pasajeros_singles}") #shape[0] --> solo devuelve nº filas

#Filas de pasajeros que sobrevivieron
print("**********************EJERCICIO 10**********************")
sobreviviventes = titanic.loc[titanic["Survived"] == 1].shape[0]
print(f"pasajeros que sobrevivieron: {sobreviviventes}")

#filas de hombres que sobrevivieron
print("**********************EJERCICIO 11**********************")
sobreviviventes_h = titanic.loc[(titanic["Survived"] == 1) & (titanic["Sex"] == "male")].shape[0]
print(f"Hombres que sobrevivieron: {sobreviviventes_h}")

#Número de columnas y filas
print("**********************EJERCICIO 12**********************")
num_col_row = titanic[["Age","Sex"]].shape
print(f"Número de columnas y filas: {num_col_row}")

#Pasajeros con cabina 2 y 3
print("**********************EJERCICIO 13**********************")
pasajeros_cabina_2_3 = titanic[(titanic['Pclass']==2) | (titanic['Pclass']==3)].head()
print(f"Pasajeros con cabina 2 y 3: {pasajeros_cabina_2_3}")

#Pasajeros con edad != 0
print("**********************EJERCICIO 14**********************")
pasajeros_edad_notnull =titanic[titanic['Age'].notna()].describe()
print(f"Pasajeros con edad != 0: {pasajeros_edad_notnull}")
pasajeros_edad_notnull.shape

#Pasajeros entre las filas 15 - 30 y columnas 4-7
print("**********************EJERCICIO 15**********************")
pasajeros_filas_col= titanic.iloc[14:30,3:7]
print(f"Pasajeros entre las filas 15 - 30 y columnas 4-7: {pasajeros_filas_col}")

#Filas compredidas entre 15 - 30 y columnas entre 1-4
print("**********************EJERCICIO 16**********************")
pasajeros_filas_col2= titanic.iloc[14:30,0:4]
print(f"Filas compredidas entre 15 - 30 y columnas entre 1-4: {pasajeros_filas_col2}")

#Edad pasajeros Titanic (20 primeros)
print("**********************EJERCICIO 17**********************")
edad = titanic["Age"]
print(f"Edad pasajeros Titanic (20 primeros): {edad.head(20)}")

#Calcula la edad media de los pasajeros del Titanic
print("**********************EJERCICIO 18**********************")
edad_media=titanic["Age"].mean()
print(f"Edad media de los pasajeros del Titanic{edad_media}")

#Valor medio de edad y precio del billete de los pasajeros
print("**********************EJERCICIO 19**********************")
media_edad = titanic["Age"].mean()
precio_medio = titanic['Fare'].mean()
print(f'Edad media: {edad_media}')
print(f'Precio medio del billete: {precio_medio}')

#valor medio, promedio mínimo, máximo, percentiles de la Edad y el precio del billete
print("**********************EJERCICIO 20**********************")
edad_media = titanic['Age'].mean()
edad_minima = titanic['Age'].min()
edad_maxima = titanic['Age'].max()
suma_edades = edad_media + edad_minima + edad_maxima
promedio = suma_edades/3
percentiles_edad = titanic['Age'].quantile([0.25,0.75])  # 25%, 75%
print(f'Valor medio: {edad_media}, mínimo: {edad_minima} y máximo es: {edad_maxima}')
print(f'El promedio es: {promedio}')
print(f'Percentiles de la Edad:\n{percentiles_edad}')


precio_media = titanic['Fare'].mean()
precio_minima = titanic['Fare'].min()
precio_maxima = titanic['Fare'].max()
suma_precios = precio_media + precio_minima + precio_maxima
promedio = suma_precios / 3
percentiles_edad = titanic['Fare'].quantile([0.25,0.75])  # 25%, 75%

print(f'Valor medio: {precio_media}, mínimo: {precio_minima} y máximo es: {precio_maxima}')
print(f'El promedio del precio es: {promedio}')
print(f'Percentiles del precio:\n{percentiles_edad}')

#Calcular la media de supervivientes del barco.
print("**********************EJERCICIO 21**********************")
media_superviviente = titanic['Survived'].mean()
print(f"Media de supervivientes: {media_superviviente}")

#Calcular el valor medio, promedio, mínimo, máximo y percentiles de los supervivientes.
print("**********************EJERCICIO 22**********************")
supervivientes_media = titanic['Survived'].mean()
supervivientes_minima = titanic['Survived'].min()
supervivientes_maxima = titanic['Survived'].max()
suma_supervivientes = precio_media + precio_minima + precio_maxima
promedio = suma_supervivientes / 3
percentiles_supervivientes = titanic['Survived'].quantile([0.25,0.75])  # 25%, 75%

print(f'Valor medio: {supervivientes_media}, mínimo: {supervivientes_minima} y máximo es: {supervivientes_maxima}')
print(f'El promedio de los supervivientes es: {promedio}')
print(f'Percentiles de los supervivientes es:\n{percentiles_supervivientes}')

#Muestra la edad media agrupadas por el sexo (de esta forma sabremos la media de los hombres y mujeres).
print("**********************EJERCICIO 23**********************")
edad_media = titanic[['Sex','Age']].groupby("Sex").mean()
print(f"La media de los hombres y de las mujeres es de: {edad_media} ")
#Aplica el groupby por sexo pero para todas las columnas
print("**********************EJERCICIO 24**********************")
agrupar = titanic.groupby('Sex').mean(numeric_only=True)
print(f"Agrupando por sexo: \n {agrupar}")
#Precio medio de la tarifa de la entrada para cada una de las combinaciones de sexo y clase de cabina
print("**********************EJERCICIO 25**********************")
precio_medio = titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
print(f"Precio medio de la tarifa de la entrada: {precio_medio}")
#Número de pasajero en cada una de las clases de cabina (Pclass)
print("**********************EJERCICIO 26**********************")
c_pasajeros = titanic['Pclass'].value_counts()
print(f"Número de pasajeros en cada cabina: {c_pasajeros}")

#número de pasajeros en cada una de las clases de cabina
print("**********************EJERCICIO 27**********************")
n_pasajeros = titanic.groupby("Pclass").size()
total_pasajeros = titanic.groupby("Pclass").count()
print(f"Número pasajeros por cabina: {n_pasajeros}")
print(f"Total pasajeros por cabina: {total_pasajeros}")

#Calcula el número de cabinas (Cabin)
print("**********************EJERCICIO 28**********************")
frecuencia_cabina = titanic.groupby(["Pclass","Cabin"])["Survived"].value_counts()
entradas_cabina_NONULL = titanic.groupby(["Pclass","Cabin"])["Survived"].count()
total_cabina = titanic.groupby(["Pclass","Cabin"])["Survived"].size()
print(f"Frecuencia de cabinas: {frecuencia_cabina}")
print(f"Entradas de cabinas (no nulas): {entradas_cabina_NONULL}")
print(f"Total de cabinas: {total_cabina}")

#Ordena los datos del Titanic según la edad de los pasajeros
print("**********************EJERCICIO 29**********************")
titanic_ordenado = titanic.sort_values(by='Age').head(15)
print(titanic_ordenado)

#Ordena los datos del Titanic según la edad de los pasajeros
print("**********************EJERCICIO 30**********************")
titanic_ordenado_descendente = titanic.sort_values(by='Age').tail(15)
print(titanic_ordenado_descendente)
```
