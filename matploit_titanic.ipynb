**Gráfica en barras que representa el número de supervivientes y fallecidos**

```
# Importar la librería
import matplotlib.pyplot as plt
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv('titanic_train (5).csv')

#Cuenta cuántos pasajeros sobrevivieron
survival_count = df['Survived'].value_counts()
print(survival_count)

#Kind es el tipo de gráfico (pie/circular ó hist/histograma ó bar/barras)
survival_count.plot(kind='bar', color=['red', 'green'])

plt.title('Representación Gráfica Supervivientes vs Fallecidos')
plt.xlabel('Supervivencia')
plt.ylabel('Cantidad')
#Editar números de los ejes
plt.xticks(ticks=[0, 1], labels=['Fallecidos', 'Supervivientes'], rotation=0)
#Muestro el histograma
plt.show()
```

**Histograma con edades de los pasajeros**

```
edades = df['Age'].value_counts()
#Creo el diagrama
edades.plot(kind="hist")
#Añado las etiquetas y el título
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
#Muestro el histograma
plt.show()
```

**Gráfico de barras de las personas de cada clase**

```
n_pasajeros = df.groupby('Pclass')['PassengerId'].count()
#Creo el diagrama
n_pasajeros.plot(kind="bar")
#Añado las etiquetas y el título
plt.title('Pasajeros por cada clase')
plt.xlabel('clase')
plt.ylabel('Pasajeros')
#Muestro el histograma
plt.show()
```

**Número de Agrupación de clase y supervivientes**

```
agrupados = df.groupby(['Pclass','Survived']).size()
#Creo el diagrama
agrupados.plot(kind="bar")
#Añado las etiquetas y el título
plt.title('Agrupación de supervivientes')
plt.xlabel('Clase')
plt.ylabel('Supervivientes')
#Muestro el histograma
plt.show()
```

**Acomulación de supervivientes por clase**

```
agrupados = df.groupby(['Pclass', 'Survived']).size()
#Creo el diagrama
agrupados.plot(kind="bar",stacked = True,color=['red', 'green']) #Stacked: agrupa por valores
#Añado las etiquetas y el título
plt.title('Acomulación de superviviente por clase')
plt.xlabel('Clase')
plt.ylabel('Número de Pasajeros')
#Muestro el histograma
plt.show()
```

**Histograma de edades y frecuencia**

```
edades = df['Age']
#Creo el diagrama
edades.plot(kind="hist",bins=40,color="pink") #bins: Divide los datos en 40 barras.
#Añado las etiquetas y el título
plt.title('Distribución Pasajeros por Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
#Muestro el histograma
plt.show()
```
