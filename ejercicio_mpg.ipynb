```
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('mpg.csv')
#Muestra los 10 priemros registros
df.head(10)
#Resumen estadístico
#df.describe()
```

```
#Coches con consumo de combustible > 25
print('---------------Ejercicio 5----------------')
mayor_mpg = df[df['mpg'] > 25]
print(mayor_mpg)
#Agrupar por n de cilindros y promedio de mpg
print('---------------Ejercicio 6----------------')
agrupados = df.groupby('cylinders')['mpg'].mean()
print(agrupados)
#convierte millas a kilómetros por litro
print('---------------Ejercicio 7----------------')
df['kpl'] = df['mpg'] * 0.425144
print(df['kpl'])
#Filtro coches con horsepower > 100 y mpg > 20
print('---------------Ejercicio 8----------------')
filtered_df = df[(df['horsepower'] > 100) & (df['mpg'] > 20)]
print(filtered_df)
#Histograma de la columna mpg.
print('---------------Ejercicio 9----------------')
plt.hist(df['mpg'], bins=20, edgecolor='black')
plt.title('Distribución de MPG')
plt.xlabel('MPG')
plt.ylabel('Frecuencia')
plt.show()
#Diagrama de dispersión (scatter plot) de mpg contra horsepower.
print('---------------Ejercicio 10----------------')
g_dispersion = plt.scatter(x = df['horsepower'], y = df['mpg']) #dibuja puntos
# Etiquetas y título
plt.title('Diagrama de dispersión: mpg vs horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Miles por Galón (MPG)')
# Mostrar el gráfico
plt.show()
#Diagrama de barras que muestre el numero de coches por cada numero de cilindros.
print('---------------Ejercicio 11----------------')
df['cylinders'].value_counts().plot(kind='bar')
plt.title('Número de Autos por Cilindros')
plt.xlabel('Cilindros')
plt.ylabel('Número de Autos')
plt.show()
#Diagrama de sectores (pie chart) que muestre la proporción de coches por origen (origin)
print('---------------Ejercicio 12----------------')
df['origin'] = df['cylinders'].value_counts()
df['origin'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporción de Autos por Origen')
plt.ylabel('')
plt.show()
# gráfico de dispersión (scatter)de las columnas mpg y kpl
print('---------------Ejercicio 13----------------')

plt.title('Comparación entre MPG y KPL')
plt.xlabel('MPG – Millas por Galón')
plt.ylabel('KPL – Kilómetros por Litro')
plt.show()
```
